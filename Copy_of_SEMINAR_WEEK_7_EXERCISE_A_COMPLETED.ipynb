## Handle Categorical Encoding

### Subtask:
Propose and apply appropriate encoding strategies for the 'experience', 'employment_type', and 'company_size' columns, considering their ordinal or nominal nature, to prepare them for modeling.


### Encoding Strategy Explanation

To prepare the categorical features for potential modeling, we will apply different encoding strategies based on their nature:

1.  **`experience_level`**: This is an ordinal variable with a clear hierarchy. We will map 'EN' (Entry-level) to 0, 'MI' (Mid-level) to 1, 'SE' (Senior-level) to 2, and 'EX' (Executive-level) to 3. This preserves the inherent order.

2.  **`company_size`**: This is also an ordinal variable. We will map 'S' (Small) to 0, 'M' (Medium) to 1, and 'L' (Large) to 2, maintaining its ordered relationship.

3.  **`employment_type`**: This is a nominal variable, meaning there is no inherent order between its categories. For such variables, one-hot encoding is appropriate. This will create new binary columns for each unique employment type, preventing the model from inferring a false sense of order or scale. We will drop the first dummy variable to avoid multicollinearity.

**Reasoning**:
Now that the encoding strategies have been explained, I will apply them by performing ordinal encoding for 'experience_level' and 'company_size' and one-hot encoding for 'employment_type', then display the head of the DataFrame to verify the transformations.



```python
import pandas as pd

# 1. Ordinal encoding for experience_level
experience_mapping = {
    'EN': 0, # Entry-level
    'MI': 1, # Mid-level
    'SE': 2, # Senior-level
    'EX': 3  # Executive-level
}
df['experience_level_encoded'] = df['experience_level'].map(experience_mapping)

# 2. Ordinal encoding for company_size
company_size_mapping = {
    'S': 0, # Small
    'M': 1, # Medium
    'L': 2  # Large
}
df['company_size_encoded'] = df['company_size'].map(company_size_mapping)

# 3. One-hot encoding for employment_type
df = pd.get_dummies(df, columns=['employment_type'], prefix='employment_type', drop_first=True)

# Display the head of the DataFrame with the newly encoded columns
print("DataFrame head with encoded columns:")
print(df.head())
```
